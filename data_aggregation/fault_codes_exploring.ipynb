### Import the dataset

```python
#import the data:
#make sure the path on the line below corresponds to the path where you put your dataset.
fault_codes = pd.read_csv('/content/drive/MyDrive/CS190/Project1/Data/OneHotEncodingData/all_faultCodes.csv')
fault_codes
```

### Perform One Hot Encoding in the chategorical variables

```python
FaultCodeDummy = pd.get_dummies(fault_codes, columns = ['ErrorMessage','ErrorType'])
FaultCodeDummy
```

### Drop not-needed columns 

```python
import pandas as pd

FaultCodeDummy.drop(columns=['TimeStamp',	'Date',	'ErrorID',	'OEM_StatusCode'])
```

### Aggregate the data by column 'TurbineName', 'interval_time'



```python
import pandas as pd


# Group the data by time and turbine name, and then sum the one-hot encoded columns
agg_data = FaultCodeDummy.groupby(['TurbineName', 'interval_time']).sum()

# Print the aggregated data
agg_data
```

### Make the dataset a csv for exploration in R

```python
from google.colab import files
agg_data.to_csv('agg_data.csv', encoding = 'utf-8-sig') 
files.download('agg_data.csv')
```
